```python
# Example of Vector
import numpy as np

# Creating a 1D list (Horizontal List)(Row)
list1 = [1,2,3]

# Creating a 1D list (Vertical List)(Column)
list2 = [[10],
        [20],
        [30]]

# Creating a vector1
# vector as row
vector1 = np.array(list1)

# Creating a vector2
# vector as a column
vector2 = np.array(list2)

print('Horizontal Vector :', vector1)
print('Vertical Vector :\n', vector2)
```

```python
# Importing important libraries
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
```

```python
# To open an image in python, open() function in Image module of PIL library is used.
# Similarly, matplotlib is used for reading and writing images.
# matplotlib uses 2 methods - imread() : to read the images, imshow() : to display the images.
# PIL - Python Imaging Library
```

```python
# Please give the appropriate path of the image you want to analyze.
# Or keep that image in the same folder where your Jupyter Notebook is.
```

```python
#  Importing the image data
img = np.array(Image.open('Tiger.jpeg'))
plt.figure(figsize = (8, 8))
plt.imshow(img)
```

```python
# Getting the details of the image
print('Number of dimensions :', img.ndim)
print('Image Shape :', img.shape)
print('Dtype :', img.dtype)
print(img[200, 600]) # This will give pixel value at RGB
print(img[:, :, 1].max())
```

```python
# Saving an image in n-dimensional array
path = 'Tiger_1.jpeg'
pil_img = Image.fromarray(img)
pil_img.save(path)
```

```python
# Getting 1D View of Data (Flattening of an array)
# Find the difference between np.ravel() and np.flatten()

# Recoding the shape
arr = np.array(img)
shape = arr.shape
print('Shape of the image array :', shape)

# Making it 1D
flat_arr = arr.ravel()
print('Flattened array :', flat_arr)
print('Shape of flattened array :', flat_arr.shape)

# Converting it back to aa matrix
vector = np.matrix(flat_arr)
print('Matrix :', vector)
print('Shape of matrix :', vector.shape)
```

```python
# Making some changes
vector[:, ::10] = 128

# Reform a numpy array of original shape
arr2 = np.asarray(vector).reshape(shape)

# Making a PIL Image
img2 = Image.fromarray(arr2, 'RGB')
img2.show()
```

```python
# Histogram for pixel densities
img = np.array(Image.open('Tiger.jpeg'))
img_flat = img.flatten()
plt.hist(img_flat, bins = 50, range = [0, 256])
plt.title('Number of pixels in each intensity value')
plt.xlabel('Intensity')
plt.ylabel('Number of Pixels')
plt.show()
```

```python
# The above graph will give you an idea about the quality of an image.
```

```python
# Flipping a given image
fig = plt.figure(figsize = (10, 10))
fig.add_subplot(2,1,1)
plt.imshow(np.flipud(img))
fig.add_subplot(2,1,2)
plt.imshow(np.fliplr(img))
```

```python
# Trimming the given image
fig = plt.figure(figsize = (10, 10))
fig.add_subplot(1,2,1)
plt.imshow(img)
plt.title('Original')
img0 = img[128:-128, 128:-128, :]
fig.add_subplot(1,2,2)
plt.imshow(img0)
plt.title('Trimmed')
plt.show()
```

```python
# Padding Black Spaces
img = np.array(Image.open('Tiger.jpeg'))
img_grey = img.sum(2)/(255*3) # Scaling the color intensities
img0 = img_grey.copy()
img0 = np.pad(img0, ((100, 100), (100, 100)), mode = 'constant')
plt.imshow(img0)
```

```python
# Negative Image
fig = plt.figure(figsize = (10, 10))
img_grey = (255*3) - img_grey
fig.add_subplot(1,2,1)
plt.imshow(img_grey)
plt.title('Negative of Grey Image')
fig.add_subplot(1,2,2)
plt.imshow(img)
plt.title('Actual Image')
plt.show()
```
